# 00 - Configuración inicial del proyecto

Este notebook tiene como objetivo validar la configuración inicial del proyecto de EDA de ventas de Olímpica.

En esta sección se revisa:

- La correcta ubicación del proyecto.
- La existencia de las carpetas principales.
- La disponibilidad del archivo original `Ventas_1.csv`.
- La instalación de librerías necesarias.
- La configuración general para trabajar los notebooks del EDA.

```python
# ===============================
# Imports principales
# ===============================

from pathlib import Path
import sys
import os

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import missingno as msno

print("Librerías cargadas correctamente.")
```

```python
# ===============================
# Configuración de rutas
# ===============================

# Ruta actual del notebook
current_path = Path.cwd()

print("Ruta actual:")
print(current_path)
```

```python
# ===============================
# Definir raíz del proyecto
# ===============================

PROJECT_ROOT = Path.cwd().parent

DATA_DIR = PROJECT_ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
INTERIM_DIR = DATA_DIR / "interim"
PROCESSED_DIR = DATA_DIR / "processed"

REPORTS_DIR = PROJECT_ROOT / "reports"
FIGURES_DIR = REPORTS_DIR / "figures"
TABLES_DIR = REPORTS_DIR / "tables"

SRC_DIR = PROJECT_ROOT / "src"

RAW_FILE = RAW_DIR / "Ventas_1.csv"

print("Raíz del proyecto:")
print(PROJECT_ROOT)

print("\nArchivo original:")
print(RAW_FILE)
```

```python
# ===============================
# Validación de carpetas principales
# ===============================

folders = {
    "PROJECT_ROOT": PROJECT_ROOT,
    "DATA_DIR": DATA_DIR,
    "RAW_DIR": RAW_DIR,
    "INTERIM_DIR": INTERIM_DIR,
    "PROCESSED_DIR": PROCESSED_DIR,
    "REPORTS_DIR": REPORTS_DIR,
    "FIGURES_DIR": FIGURES_DIR,
    "TABLES_DIR": TABLES_DIR,
    "SRC_DIR": SRC_DIR,
}

for name, path in folders.items():
    status = "Existe" if path.exists() else "No existe"
    print(f"{name:15} -> {status:10} | {path}")
```

```python
# ===============================
# Validar archivo original
# ===============================

if RAW_FILE.exists():
    print("Archivo Ventas_1.csv encontrado correctamente.")
    print(f"Ruta: {RAW_FILE}")
    print(f"Tamaño aproximado: {RAW_FILE.stat().st_size / (1024**2):.2f} MB")
else:
    print("No se encontró el archivo Ventas_1.csv.")
    print("Revisa que esté ubicado en data/raw/")
```

```python
# ===============================
# Agregar src al path de Python
# ===============================

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print("Proyecto agregado al path de Python.")
print(sys.path[-3:])
```

```python
# ===============================
# Configuración visual
# ===============================

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", "{:,.2f}".format)

sns.set_theme(style="whitegrid", context="notebook")

plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 11

print("Configuración visual aplicada.")
```

```python
# ===============================
# Lectura rápida de prueba
# ===============================

df_test = pd.read_csv(
    RAW_FILE,
    nrows=5,
    dtype={
        "CATEG": str,
        "PLU_SAP": str,
        "FACTURA": str,
        "PDV": str,
        "OFERTA_ID": str
    }
)

df_test
```
